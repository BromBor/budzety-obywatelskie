```python
from bud_lib import *
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
from sklearn.mixture import GaussianMixture
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
tqdm.pandas()
```

```python
data = pd.read_csv('budzet_dataset.csv')
data = data.dropna()
size = data.shape[0]
sentences = data["sentence"].head(size).values
```

```python
embeddings = embedd_list(sentences)
```

```python
# list of tags per category
tagi = pd.read_csv('tags.csv')
NATURA = tagi['NATURA'].dropna().values
SPOŁECZEŃSTWO = tagi['SPOŁECZEŃSTWO'].dropna().values
INFRASTRUKTURA = tagi['INFRASTRUKTURA'].dropna().values
EDUKACJA = tagi['EDUKACJA'].dropna().values
# transposed embeddings of tags
NATURA_E = np.transpose(embedd_list(NATURA))
SPOŁECZEŃSTWO_E = np.transpose(embedd_list(SPOŁECZEŃSTWO))
INFRASTRUKTURA_E = np.transpose(embedd_list(INFRASTRUKTURA))
EDUKACJA_E = np.transpose(embedd_list(EDUKACJA))
```

```python
# lists of matrix products for all sentences x tag groups
NATURA_l = []
SPOŁECZEŃSTWO_l = []
INFRASTRUKTURA_l = []
EDUKACJA_l = []
for e in embeddings:
    NATURA_l.append(np.matmul(e, NATURA_E ))
    SPOŁECZEŃSTWO_l.append(np.matmul(e, SPOŁECZEŃSTWO_E ))
    INFRASTRUKTURA_l.append(np.matmul(e, INFRASTRUKTURA_E ))
    EDUKACJA_l.append(np.matmul(e, EDUKACJA_E ))
```

```python
tags = {"NATURA" : NATURA, "SPOŁECZEŃSTWO" : SPOŁECZEŃSTWO, "EDUKACJA" : EDUKACJA, "INFRASTRUKTURA" : INFRASTRUKTURA}
tags_val = {"NATURA" : NATURA_l, "SPOŁECZEŃSTWO" : SPOŁECZEŃSTWO_l, "EDUKACJA" : EDUKACJA_l, "INFRASTRUKTURA" : INFRASTRUKTURA_l}

kategorie = []
tag = []
for i in range(size):
    # find best fitting tag for each category
    vals = {max(NATURA_l[i]): "NATURA", max(SPOŁECZEŃSTWO_l[i]): "SPOŁECZEŃSTWO", max(INFRASTRUKTURA_l[i]): "INFRASTRUKTURA",max(EDUKACJA_l[i]): "EDUKACJA"}
    # find best fitting tag of all categories -> pick category
    cat = max(vals)
    kategorie.append(vals.get(cat))
    # add best fitting tag
    p = np.argmax(tags_val.get(vals.get(cat))[i])
    tg = tags.get(vals.get(cat))
    tag.append(tg[p])
```

```python
out = data[['rok', 'miasto', 'nazwa','sentence']].head(size).copy()
out['kategorie'] = kategorie
out['tag'] = tag
out.head(100)
```
